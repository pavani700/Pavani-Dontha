```python
data = pd.read_csv('/content/sample_data/Data.csv')
data.head()
# /content/drive/My Drive/Colab Notebook/learnbay project/
```

```python
# Removing policy_id feature
data = data.drop('policy_id',axis=1)
data.shape
data.head()
```

Encoding the categorical features

```python
bin_category = ['is_esc','is_adjustable_steering','is_tpms','is_parking_sensors','is_parking_camera',
                      'is_front_fog_lights','is_rear_window_wiper','is_rear_window_washer','is_rear_window_defogger',
                      'is_brake_assist','is_power_door_locks','is_central_locking','is_power_steering',
                      'is_driver_seat_height_adjustable','is_day_night_rear_view_mirror','is_ecw','is_speed_alert']
```

```python
for cat in bin_category:
    data[cat] = data[cat].map(lambda x: 1 if x=='Yes' else 0)
```

```python
data['rear_brakes_type'] = data['rear_brakes_type'].map(lambda x: 1 if x=='Drum' else 0)
data['transmission_type'] = data['transmission_type'].map(lambda x: 1 if x=='Manual' else 0)
```

```python
pd.set_option('display.max_columns',None)
data.head()
```

```python
# Using label encoder for non-binary categorical features
from sklearn.preprocessing import LabelEncoder
def label_encoder(feat):
    le = LabelEncoder()
    le.fit(feat)
    print(feat.name, le.classes_)
    print()
    return le.transform(feat)
```

```python
label_catogery = ['area_cluster','segment','model','fuel_type','max_torque','max_power','engine_type','steering_type']
for category in label_catogery:
    data[category] = label_encoder(data[category])
```

```python
data.info()
```

```python
x = data.copy()
x = x.drop(['is_claim','policy_tenure','age_of_car','age_of_policyholder'],axis=1)  #Predictor variable assigned to x
y = pd.DataFrame(data['is_claim'],columns = ['is_claim'])    # Assigning target variable to y
print(x.shape, y.shape)
x.info()
```

```python
# Checking correlation between features
from scipy.stats import pearsonr
correlations = {}
for col in x.columns.to_list():
    corr,_ = pearsonr(x[col], y['is_claim'])
    correlations[col] = corr
sorted(correlations.items(),key=lambda x: x[1],reverse=True)
```

```python
# Outlier detection
def boxplots(col):
    sns.boxplot(x[col],orient='h')
    plt.xlabel(col)
    plt.show()

for i in list(x.select_dtypes(exclude=['object']).columns)[0:]:
    boxplots(i)
```

```python
# Check the number of outliers per columns
Q1 = x.quantile(0.25)
Q3 = x.quantile(0.75)
IQR = Q3 - Q1
UL = Q3 + 1.5 * IQR
LL = Q1 - 1.5 * IQR
```

```python
((x>UL) | (x < LL)).sum()
```

```python
# Assigning null values to the outliers
x[((x>UL) | (x < LL))] = np.NaN
```

```python
((x>UL) | (x < LL)).sum()
```

```python
# Applying capping method by using KNN imputation to replace the nullified outliers
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
x_impute = pd.DataFrame(imputer.fit_transform(x), columns=x.columns)
x_impute
```

```python
# Mutual information
from sklearn.feature_selection import mutual_info_regression
mutual_score = mutual_info_regression(x_impute.select_dtypes(exclude=['object']), y, random_state=42)
mutual_score = pd.Series(mutual_score, name='Mutual Score', index= x_impute.columns)
mutual_score = mutual_score.sort_values(ascending=False)
print('\nMutual Information Scores:')
print(mutual_score)
```

Removing features with zero variance to remove unwanted features

```python
# Check for zero variance
from sklearn.feature_selection import VarianceThreshold

# Remove features with zero variance
sel = VarianceThreshold(threshold=0)
X_reduced = sel.fit_transform(x_impute)

#Keep track of retained columns
retained_columns = x_impute.columns[sel.get_support()]

#Convert the result back to a DataFrame
X_reduced = pd.DataFrame(X_reduced, columns=retained_columns)
```

```python
X_reduced.shape
```

# Feature scaling using StandardScalar

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = pd.DataFrame(scaler.fit_transform(X_reduced), columns=X_reduced.columns)
x_scaled.describe()
```

```python
x_scaled = pd.concat([x_scaled,data[['policy_tenure','age_of_car','age_of_policyholder']]],axis=1)
x_scaled.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

Re-sampling the data to address imbalance data and splitting the dataset for trainiing and testing

```python

from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV, train_test_split

# Perform oversampling using SMOTE
smote = SMOTE(random_state=0)
x_res, y_res = smote.fit_resample(x_scaled, y)
# Split the resampled data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

```

# Random Forest

```python
 # Define the model
rf = RandomForestClassifier(random_state=0)

# Define the parameter grid
param_grid = {
    'n_estimators': [100,200]
}

# Perform grid search
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1)
grid_search.fit(x_train, y_train)

# Best parameters
print("Best parameters found: ", grid_search.best_params_)
print("Best F1 Score: ", grid_search.best_score_)

# Predict and evaluate
y_train_pred = grid_search.predict(x_train)
y_test_pred = grid_search.predict(x_test)

print("***************classification_report*********************")
print("***************Training*********************")
print(classification_report(y_train, y_train_pred))
print("*******************************")
print("***************Testing*********************")
print(classification_report(y_test, y_test_pred))
print()
print("***************accuracy_score*********************")
print(accuracy_score(y_train, y_train_pred))
print("*******************************")
print(accuracy_score(y_test, y_test_pred))
```

Random forest Confusion matrix

```python
cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot = True, fmt = 'd', cmap = 'Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Testing Confusion Matrix')
plt.show()
```

ROC- curve for Random Forest

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score


y_pred_prob = grid_search.predict_proba(x_test)[:, 1]

# Computing ROC curve and ROC area for each class
fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)
roc_auc = auc(fpr, tpr)

# Plotting the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# Define the parameter grid
param_grid = {
    'n_neighbors': [3,5,10]
}

# Initialize the KNeighborsClassifier
knn = KNeighborsClassifier()

# Perform grid search with cross-validation
grid_search = GridSearchCV(knn, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

# Get the best estimator
best_knn = grid_search.best_estimator_

# Predict using the best estimator
y_pred_train = best_knn.predict(x_train)
y_pred = best_knn.predict(x_test)

# Print the classification reports
print("***************Classification Report*********************")
print("***************Training*********************")
print(classification_report(y_train, y_pred_train))
print("*******************************")
print("***************Testing*********************")
print(classification_report(y_test, y_pred))
print()
print()
print("***************accuracy_score*********************")
print(accuracy_score(y_train, y_pred_train))
print("*******************************")
print(accuracy_score(y_test, y_pred))
# Print the best parameters
print("Best parameters found: ", grid_search.best_params_)
```
